```python
threeshold=10

counts = df['job_title'].value_counts()

major_categories = counts[counts>=threeshold].index
major_categories
```

```python
minor_categories = df[~df['job_title'].isin(major_categories)]['job_title'].unique()
minor_categories
```

from above output i extract less frequent and more frequent job

doubts i got

- less frequent job can also give high pay slary so i cant group by thresshold by frequency

- so checking with there mean salary and there frequency

- if low frequency and low salary treting as other , keep high freuency, low frequency and high mean salary

```python
counts = df['job_title'].value_counts()
avg_salary = df.groupby('job_title')['salary_in_usd'].mean()

threshold = 2
rare_jobs = counts[counts < threshold].index

# keep only top 3 highest-pay rare jobs
k = 3
high_pay_rare_jobs = avg_salary.loc[rare_jobs].sort_values(ascending=False).head(k).index
print(high_pay_rare_jobs)
group_other_jobs = set(rare_jobs) - set(high_pay_rare_jobs)

df['job_title_grouped'] = df['job_title'].replace(list(group_other_jobs), 'Other')
```
